```python
lengths = [50]
Temp = [0.5,1.0,1.5,2.0]
for T in Temp:
  for N in lengths:
    rdf = np.loadtxt(f"/bsuhome/stephaniemccallu/scratch/mse563/LAMMPS-Polymers/slurm-batch-scripts/N{N}T{T}-T{T}.rdf",skiprows=4)
    plt.plot(rdf[:,1],rdf[:,2],label=f"N:{N} T:{T}")
plt.legend()
plt.title("Rdf of 50-mers with Harmonic bonds")
```

```python
Temp = [0.5]
for T in Temp:
  for N in lengths:
    rg = np.loadtxt(f"/bsuhome/stephaniemccallu/scratch/mse563/LAMMPS-Polymers/slurm-batch-scripts/rg_N{N}T{T}.dat",skiprows=2)
    rg_avg = np.mean(rg[500:1000,1])
    print(rg_avg)
    plt.plot(N,rg_avg,label=f"N:{N} T:{T}", marker ='*')
plt.legend(loc='right')
plt.title("N vs. Radius of Gyration Harmonic")
```
